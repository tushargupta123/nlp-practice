```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u"I will google about facebook")
```

```python
doc.text
```

```python
doc[0]
```

```python
doc[0].pos_
```

```python
doc[0].tag_
```

```python
spacy.explain('PRP')
```

```python
for word in doc:
    print(word.text,'--->',word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
from spacy import displacy
```

```python
displacy.render(doc,style='dep',jupyter=True)
```
